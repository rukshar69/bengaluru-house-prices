```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
import json; import pickle
```

```python
df = pd.read_csv("/content/drive/My Drive/Personal/Projects/bengaluru_real_estate_price_prediction/bengaluru_housing_cleaned.csv")
```

```python
df.head()
```

# One hot encoding for locations(categorical)

```python
location_one_hot_df = pd.get_dummies(df['location'])
location_one_hot_df.head()
```

```python
df1 = pd.concat([df, location_one_hot_df.drop('other', axis='columns')], axis='columns')
df1 = df1.drop('location',axis='columns')
df1.shape
```

```python
feature_df = df1.drop('price',axis='columns')
target_df = df1['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature_df,target_df,test_size=0.2,random_state=10)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
lin_reg.score(X_test,y_test) #coefficient of determination -> R^2 value
```

For the given simplified dataset, a test score or accuracy of ~86% is pretty good.

```python
# save the model for later use
with open('bengaluru_property_price_prediction_model.pickle','wb') as f:
    pickle.dump(lin_reg,f)
```

```python
#Export location and column information to a file that will be useful later on in our prediction application
columns = {
    'columns' : [col.lower() for col in feature_df.columns]
}
with open("data_columns.json","w") as f:
    f.write(json.dumps(columns))
```
